```python
#defining price labels in categories
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices']<=15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
#Apply the function
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# Task 4.7.4. Calculate the Frequency of days of the week

df_orders_products_merged['hour'].value_counts(dropna = False)
```

```python
# Periods of time labeled Most orders, Average orders, and Fewest orders.
result = []

for value in df_orders_products_merged["hour"]:
    if value == 10 or value == 11 or value == 14 or value == 15 or value == 13:
        result.append("Most orders")
    elif value == 3 or value == 4 or value == 2 or value == 5 or value == 1:
        result.append("Fewest orders")
    else:
        result.append("Average orders")
```

```python
result
```

```python
# Add the new column
df_orders_products_merged['busiest_period_of_day'] = result
```

```python
# Print the frequency of the new column
df_orders_products_merged['busiest_period_of_day'].value_counts(dropna = False)
```
