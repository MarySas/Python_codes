```python
# Task 4.8.6 Create a spending_flag
result = []

for value in ords_prods_merge["prices"]:
    if value < 10:
        result.append("Low spender")
    else:
        result.append("High spender")
```

```python
ords_prods_merge['spending_flag'] = result
```

```python
#Task 4.8.7 Create a frequency flag using median of days_since_prior_order
#First transform the data in mean, and create a column with the value
```

```python
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order']>20, 'new_column_flagname'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order']<=10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check the frequency_flag
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
#Most of the customers are frequent customers.
```
